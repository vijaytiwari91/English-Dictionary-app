# 1B: Library Book Management System

## In this project, a Library Book Management System will be implemented with the front end GUI developed using Tkinter and the backend database implemented in SQLite. It is a single window tkinter GUI application which will implement CRUD( create, read, update, delete) operations on a ‘books’ database, in SQLite. The window will contain text widgets to accept / display book title, author, year of publication, ISBN (optional) and a large text area/ list box to display the records which match the search criteria.

```python
from tkinter import*
import sqlite3
import tkinter.ttk as tkk
import tkinter.messagebox as tkmessagebox
from tkinter import ttk

root = Tk()
root.title('A libray application ')

BOOKTITLE = StringVar()
AUTHOR = StringVar()
YEAR = StringVar()
ISBN = StringVar()

def Database():
    global conn,cursor
    conn = sqlite3.connect('simple_file.db')
    cursor = conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS member (mem_id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL , booktitle TEXT , author TEXT , year INTEGER, isbn TEXT)')
   # conn.close()

def Display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute('SELECT * FROM member')
    fetch = cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
    conn.close()
    BOOKTITLE.set('')
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')
    Ll1.config(text="Check the available books!",fg="red")

def Add():
    if BOOKTITLE.get()=="" and AUTHOR.get()=="" and YEAR.get()=="" and ISBN.get()=="":
        Ll1.config(text='Please enter all fields ',fg='black')
    else:
        Database()
        cursor.execute('INSERT INTO member (booktitle, author, year, isbn) VALUES(?,?,?,?)',(BOOKTITLE.get(),AUTHOR.get(),YEAR.get(),ISBN.get(),))              
        conn.commit()
        conn.close()
        BOOKTITLE.set('')
        AUTHOR.set('')
        YEAR.set('')
        ISBN.set('')
        Ll1.config(text='record inserted',fg='red')

def Issue():
    Database()
    select_var = tree.selection()
    if (len(select_var))>=0:
        tree.delete(select_var)
        Ll1.config(text='selected book is issue',fg='green')
    else:
        Ll1.config(text='please select book from Member to issue ',fg='red')
    BOOKTITLE.set('')
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')
    conn.close()

       
def Search():
    tree.delete(*tree.get_children())
    Database()
    if str(BOOKTITLE.get())!="":
        temp="SELECT * FROM member WHERE BOOKTITLE = "+ BOOKTITLE.get()
        cursor.execute(temp)
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
        Ll1.config(text="Please see the books!",fg="green")
       
    elif str(AUTHOR.get())!="":
        author="'"+str(AUTHOR.get()).title()+"'"
        temp="SELECT * FROM member WHERE AUTHOR = "+author
        cursor.execute(temp)
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
        Ll1.config(text="Please see the books!",fg="green")
       
    elif str(YEAR.get())!="":
        year ="'"+str(YEAR.get()).title()+"'"
        temp="SELECT * FROM member WHERE YEAR = "+year
        cursor.execute(temp)
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
        Ll1.config(text="Please see the books!",fg="green")
       
    elif str(ISBN.get())!="":
        isbn ="'"+str(AUTHOR.get()).title()+"'"
        temp="SELECT * FROM member WHERE ISBN = "+isbn
        cursor.execute(temp)
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert('','end',values=(data[1],data[2], data[3],data[4]))
        Ll1.config(text="Please see the books!",fg="green")
    else:
        Ll1.config(text="Please enter at least single identity!",fg="green")
    BOOKTITLE.set('')
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')
    conn.close()

    
def delete_book():
    tree.delete(*tree.get_children())
    Database()
    if BOOKTITLE.get() =='' or AUTHOR.get()=='' or YEAR.get()=='' or ISBN.get()=='' :
        Ll1.config(text='please enter all fileds',fg='red')
    else:
        cursor.execute("DELETE FROM member WHERE BOOKTITLE = ? AND Author = ? AND Year = ? AND ISBN = ?", (BOOKTITLE.get(), AUTHOR.get(),YEAR.get(),ISBN.get()))    
        conn.commit()
        Ll1.config(text='your book is delete successfully ',fg='red')
    conn.close()
    BOOKTITLE.set('')
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')

def Exit():
    root.destroy()

# frame

Top = Frame(root,width=900, height=50,bd=8,relief='raise')
Top.pack(side=TOP)
Left = Frame(root,width=300,height=500,bd=8,relief='raise',bg='red')
Left.pack(side=LEFT)
Right = Frame(root,width=600,height=500,bd=8,relief='raise',bg='yellow')
Right.pack(side=RIGHT)
inner_upper=Frame(Left,width=200,height=250,bd=8,relief="raise")
inner_upper.pack(side=TOP)

Buttons = Frame(Left,width=300,height=100,bd=8,relief='raise')
Buttons.pack(side=BOTTOM)

# ===================================================label Widget ===================
L1=Label(Top,text="A simple Library Mansgement Application",font=("arial",10))
L1.pack()

L2=Label(inner_upper,text="Book Title:",font=("arial",14))
L3=Label(inner_upper,text="Author:",font=("arial",14))
L4=Label(inner_upper,text="Year:",font=('arial',14))
L5=Label(inner_upper,text="ISBN:",font=('arial',14))

author_name=Entry(inner_upper, textvariable=BOOKTITLE)
book_title=Entry(inner_upper, textvariable=AUTHOR)
year=Entry(inner_upper, textvariable=YEAR)
isbn=Entry(inner_upper, textvariable=ISBN)
txt_result = Label(inner_upper)
txt_result.grid(row=4, column=0)
L2.grid(row=0,column=0)
book_title.grid(row=1,column=1)
L3.grid(row=1,column=0)
author_name.grid(row=0,column=1)
L4.grid(row=2,column=0)
year.grid(row=2,column=1)
L5.grid(row=3,column=0)
isbn.grid(row=3,column=1)

Ll1=Label(Right)
Ll1.pack(side=BOTTOM)

inner_frame=Frame(Left,width=250,height=50,bd=8,relief='raise')
inner_frame.pack(side=BOTTOM)

Bb1=Button(inner_frame,text="Add",command=Add)
Bb1.grid(row=0,column=0)

Bb2=Button(inner_frame,text="Display",command=Display)
Bb2.grid(row=0,column=1)

Bb3=Button(inner_frame,text="Search",command=Search)
Bb3.grid(row=0,column=2)

Bb4=Button(inner_frame,text="Issue",command=Issue)
Bb4.grid(row=0,column=3)

Bb5=Button(inner_frame,text="Delete",command=delete_book)
Bb5.grid(row=0,column=4)

Bb6=Button(inner_frame,text="Exit",command=Exit)
Bb6.grid(row=0,column=5)

scrollbary = Scrollbar(Right, orient=VERTICAL)
scrollbarx = Scrollbar(Right,orient=HORIZONTAL)
tree = ttk.Treeview(Right,column=('Booktitle','Author','Year','Isbn'))
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)

scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM, fill=X)
tree.heading('Booktitle',text='Booktitle',anchor=W)  
tree.heading('Author',text='Author',anchor=W)
tree.heading('Year',text='Year',anchor=W)
tree.heading('Isbn',text='Isbn',anchor=W)
tree.column('#0',stretch=NO, minwidth=0, width=0)
tree.column('#1',stretch=NO, minwidth=0, width=80)
tree.column('#2',stretch=NO, minwidth=0, width=120)
tree.column('#3',stretch=NO, minwidth=0, width=80)

tree.configure(xscrollcommand=scrollbarx.set)
tree.pack()
root.mainloop()
```
